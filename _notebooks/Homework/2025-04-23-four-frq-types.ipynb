# AP CSA FRQ Tips & Free Points

---

## 1. Methods and Control Structures

**Tips & Tricks:**
- Always check if you’re iterating over an array or ArrayList—know the difference.
- Watch for off-by-one errors (`<` vs `<=`, `i+1`, etc.).
- When using `if/else`, consider if there are edge cases (empty arrays, nulls).
- Use meaningful variable names to make your logic clear.

**Free Points:**
- Proper method header: `public returnType methodName(params)`
- Correct use of loop syntax (`for`, `while`)
- Return the correct value at the end
- Handle edge cases: return early if needed

---

## 2. Class Design

**Tips & Tricks:**
- Use private instance variables and always write constructors.
- Methods often include accessors (`get`), mutators (`set`), and one custom method.
- Make sure to update fields properly in methods (don’t redeclare `int x` inside a method if it’s already a field).

**Free Points:**
- Private fields (`private int x;`)
- Correct constructor signature (`public ClassName(type param)`)
- At least one `get` method and `set` method
- Method that returns or modifies fields properly

---

## 3. Array/ArrayList

**Tips & Tricks:**
- Know the syntax differences: `array.length` vs `arrayList.size()`
- Use enhanced for loops *only* when you don't need the index
- Don’t forget `.get(i)` and `.set(i, val)` for ArrayLists

**Free Points:**
- Looping through entire array/list with proper syntax
- Accessing/modifying elements correctly
- Edge case handling: empty arrays/lists
- Returning results as needed (counts, modified list, etc.)

---

## 4. 2D Array

**Tips & Tricks:**
- Use nested loops: `for (int r = 0; r < arr.length; r++)` and `for (int c = 0; c < arr[r].length; c++)`
- Understand row-major vs column-major order
- Always check `.length` for each row in jagged arrays

**Free Points:**
- Correct nested loop structure
- Accessing `arr[r][c]` safely
- Summing/counting or modifying correctly
- Avoiding `NullPointerException` or out-of-bounds errors

---

## Free Points Checklist (For Every FRQ)

- Proper method headers and class structure  
- Semicolons at the end of statements  
- No syntax errors (`ArrayList`, `int[]`, `return`, etc.)  
- Clear logic and indentation  
- Handle edge cases (empty, null, zero-length arrays)  
- Use correct variable names and accessors  
- Match return types exactly
